### Web-scraping data from the PA department of health and converting to pandas DF


```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# list of metrics
metrics = [
    'mich-01-fetal-mortality-rate-20-weeks-gestation',
    'mich-02-infant-mortality-rate-lhi',
    'mich-03-child-and-adolescent-death-rate-aged-1-to-19',
    'mich-06-percent-of-low-risk-no-prior-birth-females-giving-birth-by-cesarean',
    'mich-07-percent-of-live-births-which-are-preterm',
    'mich-08-percent-of-pregnant-females-with-early-and-adequate-prenatal-care',
    'mich-10-percent-of-females-giving-birth-who-did-not-smoke-during-pregnancy',
    'mich-13-percent-of-mothers-with-a-healthy-weight-before-giving-birth'
]

labelsFormal = ['MICH-01', 'MICH-02', 'MICH-03', 'MICH-06', 'MICH-07', 'MICH-08', 'MICH-10', 'MICH-13']
```

```python
all_metrics = pd.DataFrame()
all_vars = []
for i, m in enumerate(metrics): 
    url = 'https://www.health.pa.gov/topics/HealthStatistics/HealthyPeople/Documents/current/county/'+m+'.aspx'
    print(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table_info = soup.find_all('script')
    data_Str = table_info[15].prettify()
    dataStart = data_Str.find('Pennsylvania')-2
    dataEnd = data_Str.find('container')-3
    dataString = data_Str[dataStart:dataEnd]
    dataList = eval(dataString)
    dataDF = pd.DataFrame(dataList).T
    varnames = ['2011-2015', '2012-2016', '2013-2017', '2014-2018', '2015-2019', 'MetGoal']
    m_specific_vars = [labelsFormal[i]+'_'+var for var in varnames]
    all_vars = all_vars+m_specific_vars
    dataDF.columns = ['County'+str(i)]+m_specific_vars
    
    all_metrics = pd.concat([all_metrics, dataDF], axis="columns")

keep_vars = ['County0']+all_vars

all_metrics = all_metrics[keep_vars]
all_metrics
```

```python
all_metrics.to_csv('PAHealth_MaternalChildHealth_20221127.csv')
```

### Hospital Data

```python
url = 'https://www.ahd.com/states/hospital_PA.html'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
table_info = soup.find(class_ = 'report')
tabledata = [i.get_text().strip() for i in table_info.find_all('td')]
tabledata[0:6]
```

```python
rows = []
for i in range(183):
    if i == 0: 
        startIndex = 0
    else:
        startIndex = startIndex+6
    row_i = tabledata[startIndex:startIndex+6]
    rows.append(row_i)
```

```python
tableData_Df = pd.DataFrame(rows)
tableData_Df.columns = ['HospitalName', 'City', 'NumberOfStaffedBeds', 'TotalDischarges', 'PatientDays', 'GrossPatientRevenue($000)']
tableData_Df
```

```python
# adjust data types
tableData_Df.NumberOfStaffedBeds = tableData_Df.NumberOfStaffedBeds.apply(lambda x: x.replace(',', ''))
tableData_Df.NumberOfStaffedBeds = tableData_Df.NumberOfStaffedBeds.apply(lambda x: int(x))

tableData_Df.TotalDischarges = tableData_Df.TotalDischarges.apply(lambda x: x.replace(',', ''))
tableData_Df.TotalDischarges = tableData_Df.TotalDischarges.apply(lambda x: int(x))

tableData_Df.PatientDays = tableData_Df.PatientDays.apply(lambda x: x.replace(',', ''))
tableData_Df.PatientDays = tableData_Df.PatientDays.apply(lambda x: int(x))

tableData_Df['GrossPatientRevenue($000)'] = tableData_Df['GrossPatientRevenue($000)'].apply(
    lambda x: x.replace(',', '').replace('$', ''))
tableData_Df['GrossPatientRevenue($000)'] = tableData_Df['GrossPatientRevenue($000)'].apply(lambda x: float(x))
```
